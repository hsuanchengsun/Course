# 畫一張圖

```python
# 畫一張圖

x = np.linspace(0, 10, 1000)
y1 = np.sin(x) 
y2 = np.cos(x) 

plt.plot(x, y1)
plt.plot(x, y2)
```

# 接力畫一張完整的圖

```python
# 畫一張圖

plt.figure(figsize=(6, 6)) #設定圖的大小是(6, 6)
plt.xkcd() #設定一個風格
plt.plot(x, y1, label = "sin(x)") #y1的名字是"sin"
plt.plot(x, y2, label = "cos(x)") #y2的名字是"cos"
plt.legend(loc='lower left') #將圖標放在左下角

plt.xlabel("X Axis") #X軸的名字是"X Axis"
plt.ylabel("Y Axis") #Y軸的名字是"Y Axis"


plt.xlim(0, 10) #設定X軸的範圍
plt.ylim(-1, 1) #設定Y軸的範圍

plt.xticks(np.arange(0, 10, 1)) #X軸分成10個部分
plt.yticks(np.arange(-1, 1, 0.5)) #Y軸分成4個部分

plt.title("Sin and Cos") #圖的標題是"Sin and Cos"
```

# Variable Length Function Arguments

```python
def ARGV(*argv):  
    for i in argv:  
        print(i)
```

```python
ARGV("Hello", "NCTS", "I", "am", "hungry.", 88)
```

```python
def ARGV1(arg1, *argv):  
    for i in argv:  
        print(i)
```

```python
ARGV1("Hello", "NCTS", "I", "am", "hungry.", 88)
```

```python
def ARGV2(arg1, *argv, arg2):  
    for i in argv:  
        print(i)
```

```python
ARGV2("Hello", "NCTS", "I", "am", "hungry.", 88)
```

# Variable Length Keyword Argument

```python
def KWARGS(**kwargs):  
    for key, value in kwargs.items(): 
        print (f"{key} == {value}") 
```

```python
KWARGS(first="N", second="C", third="T", fourth="S")
```

```python
def Four(arg1, arg2, arg3, arg4): 
    print("arg1 is ", arg1) 
    print("arg2 is ", arg2) 
    print("arg3 is ", arg3) 
    print("arg4 is ", arg4)
```

```python
args = ("N", "C", "T", "S") 
Four(*args) 
```

```python
kwargs = {"arg2" : "C", "arg1" : "N", "arg4" : "S", "arg3" : "T"} 
Four(**kwargs) 
```

```python
def Mix(*args, **kwargs):
    print(type(args))
    for v in args:
        print(v)
        
    print()

    print(type(kwargs))
    for k, v in kwargs.items():
        print(f'kwargs[{k}] = {v}')
```

```python
Mix("N", "C", first = "T", second = "S")
```

# Enumerate
