### groupby

```python
ufo_state = ufo.groupby("State")
```

```python
ufo_state
```

```python
ufo_state = ufo.groupby("State").count()
```

```python
ufo_state
```

```python
ufo_state.sort_values(by = "Time", ascending= False, inplace= True)
```

```python
ufo_state[:10].Time.plot(kind = "bar")
```

```python
ufo.groupby(["State", "City"]).count()
```

```python
ufo_SC = ufo.groupby(["State", "City"]).count()
```

### datetime

```python
ufo["datetime"] = pd.to_datetime(ufo.Time)
```

```python
ufo.head()
```

```python
import datetime as dt
```

```python
ufo["year"] = ufo.datetime.dt.year
```

```python
ufo["month"] = ufo.datetime.dt.month
```

```python
ufo["hour"] = ufo.datetime.dt.hour
```

```python
ufo["day_of_the_week"] = ufo.datetime.dt.weekday_name
```

```python
ufo_hour = ufo.groupby("hour").count()
```
