# matplotlib 相關功能

```python
x = [1, 2, 3, 4]
y1 = [7, 3, 9, 2]
y2 = [5, 0 ,2 ,3]
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize=(5, 5), facecolor="b")
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")
plt.legend(loc='best')
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc='best')

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.title("Title")
```

```python
import numpy as np
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc='best')

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xlim(-2, 12)
plt.ylim(-20, 10)
plt.xticks(np.arange(0, 15, 5))
plt.yticks(np.arange(-15, 10, 5))

plt.title("Title")
```

```python
plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'b:')
```
